```python
import numpy as np
```

```python
def average_light_fun(mu_arr, light_arr):
    ave = np.zeros(np.unique(light_arr).shape[0])
    lights_ave = np.unique(light_arr)
    count_arr = np.zeros(np.unique(light_arr).shape[0])

    print(ave)
    print(lights_ave)
    print(count_arr)
    for i in range(len(light_arr)):
        count_arr[np.where(lights_ave == light_arr[i])] += 1
        ave[np.where(lights_ave == light_arr[i])] += mu_arr[i]
    return lights_ave, ave / count_arr
    
```
