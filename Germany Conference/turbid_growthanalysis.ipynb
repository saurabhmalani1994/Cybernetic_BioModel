```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.integrate import odeint, solve_ivp
from scipy.optimize import minimize
from scipy.interpolate import interp1d
import math

reactor_name = 'M5'

filename = '/home/smalani/Cybernetic/Germany Conference/training_data/Analyzed Data Oct 15.xlsx'
reactor_deact = pd.read_excel(filename, sheet_name=reactor_name)
filename = '/home/smalani/Cybernetic/Germany Conference/training_data/Analyzed Data Oct 17.xlsx'
reactor_act = pd.read_excel(filename, sheet_name=reactor_name)

act_cutoff = 5
time_array_act = reactor_act['Time (hrs)'].values
opto_array_act = reactor_act['opt_gen_act_int'].values[time_array_act > act_cutoff]
growth_rate_array_act = reactor_act['growth_rate'].values[time_array_act > act_cutoff]
time_array_act = time_array_act[time_array_act > act_cutoff] - time_array_act[time_array_act > act_cutoff][0]


deact_cutoff = 10
time_array_deact = reactor_deact['Time (hrs)'].values
opto_array_deact = reactor_deact['opt_gen_act_int'].values[time_array_deact > deact_cutoff]
growth_rate_array_deact = reactor_deact['growth_rate'].values[time_array_deact > deact_cutoff]
time_array_deact = time_array_deact[time_array_deact > deact_cutoff] - time_array_deact[time_array_deact > deact_cutoff][0]

act_pars = (time_array_act, opto_array_act, growth_rate_array_act)
deact_pars = (time_array_deact, opto_array_deact, growth_rate_array_deact)

# define first-order plus dead-time approximation    
def fopdt(t,y,uf,Km,taum,thetam, yp0, u0):
    # arguments
    #  y      = output
    #  t      = time
    #  uf     = input linear function (for time shift)
    #  Km     = model gain
    #  taum   = model time constant
    #  thetam = model time constant
    # time-shift u
    try:
        if (t-thetam) <= 0:
            um = uf(0.0)
        else:
            um = uf(t-thetam)
    except:
        #print('Error with time extrapolation: ' + str(t))
        um = u0
    # calculate derivative
    dydt = (-(y-yp0) + Km * (um-u0))/taum
    return [dydt]

# simulate FOPDT model with x=[Km,taum,thetam]
def sim_model(x, time_array, opto_array, growth_rate_array):
    # input arguments
    Km = x[0]
    taum = x[1]
    thetam = x[2]
    y_init = [x[3]]


    # initial condition

    uf = interp1d(time_array, opto_array)
    # loop through time steps    
    sol = solve_ivp(fopdt, (time_array[0], time_array[-1]), y_init, 
        args=(uf,Km,taum,thetam, y_init[0], opto_array[0]), t_eval=time_array)

    return sol.y[0]

# define objective
def objective(x, reactor_pars, info={'Nfeval':0}):
    # time_array_act, opto_array_act, growth_rate_array_act = act_pars
    # time_array_deact, opto_array_deact, growth_rate_array_deact = deact_pars

    # simulate model
    ym = sim_model(x, *reactor_pars)
    # calculate objective
    obj = 0.0
    obj += np.sum((ym - reactor_pars[2])**2)

    if info['Nfeval']%50 == 0:
        print('{0:4d} {1: 3.6f}   {2: 3.6f}   {3: 3.6f}   {4: 3.6f}'.format(info['Nfeval'], x[0], x[1], x[2], obj))
    info['Nfeval'] += 1
    return obj


# initial guesses
x0 = np.zeros(4)
x0[0] = -0.4 # Km
x0[1] = 3.900776 # taum
x0[2] = 1.332222 # thetam
x0[3] = 0.5 # y_init

# assert False
# optimize Km, taum, thetam
solution = minimize(objective,x0,args=(deact_pars, {'Nfeval':0}),options={'maxiter':200} )

# Another way to solve: with bounds on variables
#bnds = ((0.4, 0.6), (1.0, 10.0), (0.0, 30.0))
#solution = minimize(objective,x0,bounds=bnds,method='SLSQP')
x_deact = solution.x

# show final objective
print('Final SSE Objective deact: ' + str(objective(x0, deact_pars)))

print('Kp deact: ' + str(x_deact[0]))
print('taup deact: ' + str(x_deact[1]))
print('thetap deact: ' + str(x_deact[2]))
print('y_init deact: ' + str(x_deact[3]))

x0 = np.zeros(4)
x0[0] = -0.49 # Km
x0[1] = 9 # taum
x0[2] = 3.5 # thetam
x0[3] = 0.08 # y_init

solution = minimize(objective,x_deact,args=(act_pars, {'Nfeval':0}),options={'maxiter':1000} )

x_act = solution.x

# show final objective
print('Final SSE Objective act: ' + str(objective(x0, act_pars)))

print('Kp act: ' + str(x_act[0]))
print('taup act: ' + str(x_act[1]))
print('thetap act: ' + str(x_act[2]))
print('y_init act: ' + str(x_act[3]))

# calculate model with updated parameters
ym2 = sim_model(x_deact, *deact_pars)   
# plot results
plt.figure(figsize=(10,10))
plt.subplot(2,1,1)
plt.plot(time_array_deact,growth_rate_array_deact,'kx-',linewidth=2,label='Process Data')
plt.plot(time_array_deact,ym2,'r--',linewidth=3,label='Optimized FOPDT')
plt.ylabel('Output')
plt.legend(loc='best')
plt.subplot(2,1,2)

ym2 = sim_model(x_act, *act_pars)
plt.plot(time_array_act,growth_rate_array_act,'kx-',linewidth=2,label='Process Data')
plt.plot(time_array_act,ym2,'r--',linewidth=3,label='Optimized FOPDT')
plt.ylabel('Output')
plt.legend(loc='best')
plt.show()
```
